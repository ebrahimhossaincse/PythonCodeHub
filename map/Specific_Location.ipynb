## Install the necessary packages: pip install folium

```python
import folium
from IPython.display import display

# Center the map on Dhaka, Bangladesh
map_center = [23.8103, 90.4125]
my_map = folium.Map(location=map_center, zoom_start=10)

# Add a marker for Dhaka
folium.Marker(
    [23.8103, 90.4125],
    popup='Dhaka',
    icon=folium.Icon(color='blue', icon='info-sign')
).add_to(my_map)

display(my_map)
```
