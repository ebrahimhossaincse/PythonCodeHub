```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters for multiple circles
circles = [
    {"radius": 2, "x_center": 0, "y_center": 0},
    {"radius": 3, "x_center": 4, "y_center": 4},
    {"radius": 1.5, "x_center": -3, "y_center": 3},
    {"radius": 2.5, "x_center": -4, "y_center": -4}
]

# Generate theta values for circle parameterization
t = np.linspace(0, 2 * np.pi, 1000)

# Create a Circle graph
fig, ax = plt.subplots()
for circle in circles:
    radius = circle["radius"]
    x_center = circle["x_center"]
    y_center = circle["y_center"]
    
    # Calculate x and y points for each circle
    x = x_center + radius * np.cos(t)
    y = y_center + radius * np.sin(t)
    
    # Plot each circle
    ax.plot(x, y, label=f'Center: ({x_center}, {y_center}), Radius: {radius}')

# Add title and labels
ax.set_title('Multiple Circles Graph')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.axis('equal')  # Ensures the aspect ratio is equal
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters for multiple parabolas
parabolas = [
    {"a": 1, "b": 0, "c": 0, "label": "y = x^2"},
    {"a": 0.5, "b": -1, "c": 2, "label": "y = 0.5x^2 - x + 2"},
    {"a": -0.2, "b": 1, "c": -3, "label": "y = -0.2x^2 + x - 3"},
    {"a": 1.5, "b": 0, "c": -1, "label": "y = 1.5x^2 - 1"}
]

# Generate x values
x = np.linspace(-10, 10, 1000)

# Create a Parabola graph
fig, ax = plt.subplots()
for parabola in parabolas:
    a = parabola["a"]
    b = parabola["b"]
    c = parabola["c"]
    
    # Calculate y points for each parabola
    y = a * x**2 + b * x + c
    
    # Plot each parabola
    ax.plot(x, y, label=parabola["label"])

# Add title, labels, and legend
ax.set_title('Multiple Parabolas')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.grid(True)
plt.show()
```
