```python
import matplotlib.pyplot as plt
import numpy as np
```

### Define the Circle parameters

```python
radius = float(input('Enter radius of circle: '))
x_center = float(input('Enter x center of circle: '))
y_center = float(input('Enter y center of circle: '))
```

### Generate Some Data

```python
t = np.linspace(0, 2*np.pi, 1000)
x = x_center*np.cos(t)
y = y_center*np.sin(t)
```

### Create a Circle graph

```python
fig, ax = plt.subplots()
ax.plot(x, y)

ax.set_title('Circle Graph')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```
